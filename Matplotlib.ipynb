```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[10,25,35,40,50]

plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Basic Line Plot")
plt.plot(x,y,color='blue')
plt.plot(x,y,marker='*')
plt.show()

x=[1,2,3,4,5]
y=[10,25,35,40,50]

plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Basic Line Plot")
plt.plot(x,y,color='blue')
plt.plot(x,y,linestyle='--')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,25,35,40,50]

plt.scatter(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Basic Scatter Plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,25,35,40,50]

plt.bar(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Bar Graph")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,25,35,40,50]

plt.barh(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Bar Graph (Horizontal)")
plt.show()
```

```python
x=['A','B','C','D','E']
y=[10,25,35,40,50]

plt.pie(y,labels=x)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Pie Chart")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,25,35,40,50]

plt.stem(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Stem Graph")
plt.show()
```
